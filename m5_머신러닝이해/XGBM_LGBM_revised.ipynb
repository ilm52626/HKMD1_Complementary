XGBoost는 Gradient Boosting 알고리즘을 기반으로 한 머신러닝 알고리즘입니다. "eXtreme Gradient Boosting"의 약자로, 고성능이며 정확도가 높은 예측 모델을 구축하는 데 사용됩니다. XGBoost는 트리 기반의 앙상블 학습 방법으로, 여러 개의 결정 트리를 조합하여 예측을 수행합니다.

XGBoost의 특징은 다음과 같습니다:

1. Gradient Boosting 기법: 이전 트리의 오차를 보완하는 방식으로 순차적으로 트리를 학습시킵니다. 오차에 대한 그래디언트를 계산하여 예측값을 업데이트하고, 새로운 트리를 학습시켜 오차를 줄여나갑니다.

2. 정규화 및 규제: XGBoost는 과적합을 방지하기 위해 정규화와 규제 기법을 사용합니다. 정규화는 트리의 복잡도를 제한하고 규제는 트리의 가중치를 조절하여 일반화 성능을 향상시킵니다.

3. 속도와 성능: XGBoost는 최적화된 알고리즘과 병렬 처리를 통해 빠른 속도와 높은 성능을 제공합니다. 효율적인 구현과 알고리즘의 최적화로 대용량 데이터셋에서도 뛰어난 성능을 발휘합니다.

4. 특성 중요도: XGBoost는 각 특성의 중요도를 평가할 수 있는 기능을 제공합니다. 특성의 기여도를 계산하여 모델의 해석력을 높이고, 중요한 특성을 식별하는 데 도움을 줍니다.

XGBoost는 다양한 종류의 문제에 적용될 수 있으며, 분류와 회귀 문제에서 좋은 성능을 보입니다. 파이썬에서는 XGBoost 라이브러리를 사용하여 간편하게 모델을 구축하고 예측을 수행할 수 있습니다.

##Boosting

Boosting은 약한 학습기(weak learner)를 여러 개 결합하여 강한 학습기(strong learner)를 만드는 앙상블(Ensemble) 기법입니다. 약한 학습기는 일반적으로 성능이 낮은 모델로서, 예측 오차에 가중치를 부여하여 순차적으로 학습을 진행하면서 오차를 보완해 나갑니다. 이를 통해 강한 학습기를 구성하고 예측 성능을 향상시킵니다.

XGBoost는 Gradient Boosting 알고리즘을 기반으로 한 Boosting 알고리즘의 한 종류입니다. "eXtreme Gradient Boosting"의 약자로, 고성능이며 정확도가 높은 예측 모델을 구축하는 데 사용됩니다. XGBoost는 트리 기반의 앙상블 학습 방법으로, 여러 개의 결정 트리를 조합하여 예측을 수행합니다.

XGBoost의 주요 장점은 다음과 같습니다:

빠른 수행 시간: 병렬 처리와 최적화된 알고리즘을 사용하여 빠른 속도로 학습과 예측을 수행합니다.
과적합 규제: 내부적으로 과적합 규제 기능을 제공하여 일반화 성능을 향상시킵니다.
예측 성능: 분류와 회귀 문제에서 모두 높은 예측 성능을 보이며, 다양한 옵션을 제공하여 모델을 조정할 수 있습니다.
조기 종료 기능: Early Stopping을 지원하여 최적의 학습 횟수를 자동으로 결정할 수 있습니다.
결측값 처리: 내부적으로 결측값을 처리하여 편리한 모델 구축이 가능합니다.
XGBoost는 다양한 데이터 분석 및 예측 과제에 활용되며, 파이썬에서는 XGBoost 라이브러리를 사용하여 간편하게 모델을 구축하고 예측을 수행할 수 있습니다.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

dataset = load_breast_cancer()
X_features= dataset.data
y_label = dataset.target

cancer_df = pd.DataFrame(data=X_features, columns=dataset.feature_names)
cancer_df['target']= y_label
cancer_df.head(3)
```

```python
print(dataset.target_names)
print(cancer_df['target'].value_counts())
```

```python
# 전체 데이터 중 80%는 학습용 데이터, 20%는 테스트용 데이터 추출
X_train, X_test, y_train, y_test=train_test_split(X_features, y_label,
                                         test_size=0.2, random_state=156 )
print(X_train.shape , X_test.shape)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score, roc_auc_score

def get_clf_eval(y_test , pred, pred_proba):
    confusion = confusion_matrix( y_test, pred)
    accuracy = accuracy_score(y_test , pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    f1 = f1_score(y_test,pred)
    roc_auc = roc_auc_score(y_test, pred_proba)
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f},\
    F1: {3:.4f}, AUC:{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
# 사이킷런 래퍼 XGBoost 클래스인 XGBClassifier 임포트
from xgboost import XGBClassifier

evals = [(X_test, y_test)]
xgb_wrapper = XGBClassifier(n_estimators=400, learning_rate=0.1, max_depth=3)
xgb_wrapper.fit(X_train , y_train,  early_stopping_rounds=400,eval_set=evals, eval_metric="logloss",  verbose=True)
w_preds = xgb_wrapper.predict(X_test)
w_preds_proba=xgb_wrapper.predict_proba(X_test)[:,1]
```

```python
get_clf_eval(y_test , w_preds, w_preds_proba)
```

```python
# 211번에서 311번까지 반복 동안 성능 평가 지수가 향상되지 않으므로 STOP(early_stopping_rounds=100)
from xgboost import XGBClassifier

xgb_wrapper = XGBClassifier(n_estimators=400, learning_rate=0.1, max_depth=3)
evals = [(X_test, y_test)]
xgb_wrapper.fit(X_train, y_train, early_stopping_rounds=100, eval_metric="logloss", 
                eval_set=evals, verbose=True)
ws100_preds = xgb_wrapper.predict(X_test)
ws100_preds_proba = xgb_wrapper.predict_proba(X_test)[:,1]
```

```python
get_clf_eval(y_test , ws100_preds, ws100_preds_proba)
```

```python
# early_stopping_rounds를 10으로 설정하고 재 학습. 
xgb_wrapper.fit(X_train, y_train, early_stopping_rounds=10, # 조기 중단할 수 있는 최소 반복 횟수
                eval_metric="logloss", eval_set=evals,verbose=True)

ws10_preds = xgb_wrapper.predict(X_test)
ws10_preds_proba = xgb_wrapper.predict_proba(X_test)[:,1]
get_clf_eval(y_test , ws10_preds, ws10_preds_proba)
```
