군집분석은 데이터를 유사한 특성을 가진 그룹으로 나누는 분석 방법입니다. 우리는 데이터의 패턴과 유사성을 파악하여 데이터를 의미 있는 그룹으로 분류하고자 합니다. 이렇게 분류된 그룹을 '군집'이라고 부릅니다.

군집분석은 매우 다양한 분야에서 활용될 수 있습니다. 예를 들어, 고객 세그멘테이션을 통해 비슷한 특성을 가진 고객들을 그룹화하여 효율적인 타겟 마케팅이나 개인화된 서비스 제공을 할 수 있습니다. 또는 의료 분야에서 환자들을 유사한 질병 패턴을 가진 그룹으로 나누어 질병 예측이나 치료 방법 개발에 활용할 수도 있습니다.

파이썬에서는 여러 라이브러리를 사용하여 군집분석을 수행할 수 있습니다. 가장 대표적인 라이브러리는 scikit-learn입니다. scikit-learn에서는 다양한 군집분석 알고리즘을 제공하고 있으며, 이를 활용하여 데이터를 군집화할 수 있습니다.

군집화(Clustering) 알고리즘에는 다양한 방법이 있습니다. 여기서는 K-평균(K-Means), 평균 이동(Mean Shift), 가우시안 혼합 모델(Gaussian Mixture Model, GMM), DBSCAN(Density-Based Spatial Clustering of Applications with Noise)에 대해 설명하겠습니다.

1. K-평균 (K-Means):
   - K-평균은 가장 널리 사용되는 군집화 알고리즘 중 하나입니다.
   - 동작 방식은 데이터 포인트를 K개의 그룹(군집)으로 나누는 것입니다.
   - 초기에 K개의 중심점을 랜덤하게 선택하고, 각 데이터 포인트를 가장 가까운 중심점에 할당합니다.
   - 할당된 데이터 포인트들의 평균 위치를 새로운 중심점으로 업데이트하고, 중심점이 수렴할 때까지 반복합니다.
   - 최종적으로 형성된 그룹은 각 데이터 포인트가 가장 가까운 중심점에 속하는 그룹입니다.

2. 평균 이동 (Mean Shift):
   - 평균 이동은 밀도 기반 군집화 알고리즘입니다.
   - 데이터 포인트를 중심으로 반경 내의 데이터 포인트들의 평균 위치를 계산하여 중심을 이동시킵니다.
   - 이동된 중심으로 다시 반경 내의 데이터 포인트들의 평균 위치를 계산하고, 이 과정을 반복하여 중심이 수렴할 때까지 진행합니다.
   - 최종적으로 형성된 그룹은 중심이 수렴한 지점들로 이루어진 그룹입니다.

3. 가우시안 혼합 모델 (Gaussian Mixture Model, GMM):
   - 가우시안 혼합 모델은 확률적 군집화 방법으로, 데이터가 여러 개의 가우시안 분포를 따른다고 가정합니다.
   - 각 데이터 포인트가 어떤 가우시안 분포에 속하는지 확률적으로 추정합니다.
   - EM(Expectation-Maximization) 알고리즘을 사용하여 가우시안 분포의 평균과 분산을 조정하면서 군집화를 수행합니다.
   - 최종적으로 형성된 그룹은 각 데이터 포인트가 속할 가능성이 가장 높은 가우시안 분포로 이루어진 그룹입니다.

4. DBSCAN (Density-Based Spatial Clustering of Applications with Noise):
   - DBSCAN은 데이터 포인트들의 밀도를 기반으로 군집을 형성하는 알고리즘입니다. 
   - 주어진 거리(ε) 내에 최소한 지정된 개수(MinPts) 이상의 이웃 데이터 포인트를 가지는 데이터 포인트를 핵심 포인트로 간주하고 군집을 형성합니다. 
   - 군집의 개수를 사전에 지정할 필요가 없고 다양한 모양과 크기의 군집을 탐지할 수 있습니다.

## 예시1 - 꽃 품종

K-means


```python
from sklearn.preprocessing import scale
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, 
                       columns=['sepal length','sepal width','petal length','petal width'])
df_iris.head()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', 
                max_iter=300,random_state=0).fit(df_iris)
print(kmeans.labels_)
```

```python
df_iris['cluster'] = kmeans.labels_
df_iris.head()
```

```python
df_iris['target'] = iris.target
df_iris.head()
```

```python
iris_result = df_iris.groupby(['target','cluster'])['sepal length'].count()
iris_result
```

### PCA(주성분분석) 차원축소


```python
from sklearn.preprocessing import StandardScaler
iris_scaled = StandardScaler().fit_transform(df_iris)
```

```python
# 4개의 붓꽃데이터 속성은 2차원 평면상에 적합치 않아 PCA를 이용 4개의 속성을 
# 2개 차원 축소한 뒤에 X좌표, Y좌표로 개별 데이터를 표현

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(iris_scaled.data)
pca_transformed[:5]
```

```python
df_iris['pca_x'] = pca_transformed[:,0]
df_iris['pca_y'] = pca_transformed[:,1]
df_iris.head()
```

```python
# cluster 값이 0,1,2인 경우마다 별도의 index로 추출
mark0_ind = df_iris[df_iris.cluster==0].index
mark1_ind = df_iris[df_iris.cluster==1].index
mark2_ind = df_iris[df_iris.cluster==2].index

plt.scatter(x=df_iris.loc[mark0_ind,'pca_x'], 
            y=df_iris.loc[mark0_ind,'pca_y'],marker='o')
plt.scatter(x=df_iris.loc[mark1_ind,'pca_x'], 
            y=df_iris.loc[mark1_ind,'pca_y'],marker='s')
plt.scatter(x=df_iris.loc[mark2_ind,'pca_x'], 
            y=df_iris.loc[mark2_ind,'pca_y'],marker='^')

plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('3 Clusters Visualization by 2 PCA Components')
```

## 평가지표


군집화의 평가를 위해 실루엣 분석을 사용할 수 있습니다. 실루엣 분석은 각 군집 간의 거리가 얼마나 효과적으로 분리되어 있는지를 평가하는 지표입니다.

실루엣 계수는 -1에서 1 사이의 값을 가지며, 더 큰 값일수록 군집 간 거리가 멀리 떨어져 있다는 것을 의미합니다. 0에 가까울수록 군집 간 거리가 가까워진다는 것을 나타냅니다.

실루엣 계수는 개별 데이터 포인트의 군집 내 다른 데이터 포인트와의 평균 거리(a(i))와 해당 데이터 포인트가 속하지 않은 군집 중 가장 가까운 군집과의 평균 거리(b(i))를 비교하여 계산됩니다. 이를 식으로 표현하면 s(i) = (b(i) - a(i)) / max(a(i), b(i))입니다.

실루엣 분석을 수행하기 위해 사용되는 함수는 다음과 같습니다:

- silhouette_samples: 각 데이터 포인트의 실루엣 계수를 계산하여 반환합니다. 입력으로 X feature 데이터 세트와 각 데이터 포인트의 군집 레이블 값인 labels 데이터를 사용합니다.

- silhouette_score: 전체 데이터의 실루엣 계수 값을 평균하여 반환합니다. 입력으로 X feature 데이터 세트와 각 데이터 포인트의 군집 레이블 값인 labels 데이터를 사용합니다. 즉, np.mean(silhouette_samples())와 동일한 값을 반환합니다.

실루엣 계수의 평균값과 개별 군집의 평균값의 편차가 작을수록 군집화가 좋다고 평가할 수 있습니다. 하지만 실루엣 계수만으로 군집화의 성능을 절대적으로 판단할 수는 없으며, 다른 평가 지표와 함께 고려해야 합니다.

```python
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import scale
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

iris = load_iris()
feature_names = ['sepal_length','sepal_width','petal_length','petal_width']
df_iris = pd.DataFrame(data=iris.data, columns = feature_names)
kmeans = KMeans(n_clusters=3, init='k-means++', 
                max_iter=300, random_state=0).fit(df_iris)
df_iris ['cluster'] = kmeans.labels_

score_samples = silhouette_samples(iris.data, df_iris['cluster'])
print(score_samples.shape)

df_iris['silhouette_coeff'] = score_samples

average_score = silhouette_score(iris.data, df_iris['cluster'])

print('score:{:.4f}'.format(average_score))
df_iris.head()
```

```python
df_iris.groupby('cluster')['silhouette_coeff'].mean()
```

```python
# 여러개의 클러스터링 갯수를 List로 입력 받아 각각의 실루엣 계수를 면적으로 시각화한 함수 작성
def visualize_silhouette(cluster_lists, X_features): 
    
    from sklearn.datasets import make_blobs
    from sklearn.cluster import KMeans
    from sklearn.metrics import silhouette_samples, silhouette_score

    import matplotlib.pyplot as plt
    import matplotlib.cm as cm
    import math
    
    # 입력값으로 클러스터링 갯수들을 리스트로 받아서, 각 갯수별로 클러스터링을 적용하고 실루엣 개수를 구함
    n_cols = len(cluster_lists)
    
    # plt.subplots()으로 리스트에 기재된 클러스터링 수만큼의 sub figures를 가지는 axs 생성 
    fig, axs = plt.subplots(figsize=(4*n_cols, 4), nrows=1, ncols=n_cols)
    
    # 리스트에 기재된 클러스터링 갯수들을 차례로 iteration 수행하면서 실루엣 개수 시각화
    for ind, n_cluster in enumerate(cluster_lists):
        
        # KMeans 클러스터링 수행하고, 실루엣 스코어와 개별 데이터의 실루엣 값 계산. 
        clusterer = KMeans(n_clusters = n_cluster, max_iter=500, random_state=0)
        cluster_labels = clusterer.fit_predict(X_features)
        
        sil_avg = silhouette_score(X_features, cluster_labels)
        sil_values = silhouette_samples(X_features, cluster_labels)
        
        y_lower = 10
        axs[ind].set_title('Number of Cluster : '+ str(n_cluster)+'\n' \
                          'Silhouette Score :' + str(round(sil_avg,3)) )
        axs[ind].set_xlabel("The silhouette coefficient values")
        axs[ind].set_ylabel("Cluster label")
        axs[ind].set_xlim([-0.1, 1])
        axs[ind].set_ylim([0, len(X_features) + (n_cluster + 1) * 10])
        axs[ind].set_yticks([])  # Clear the yaxis labels / ticks
        axs[ind].set_xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
        
        # 클러스터링 갯수별로 fill_betweenx( )형태의 막대 그래프 표현. 
        for i in range(n_cluster):
            ith_cluster_sil_values = sil_values[cluster_labels==i]
            ith_cluster_sil_values.sort()
            
            size_cluster_i = ith_cluster_sil_values.shape[0]
            y_upper = y_lower + size_cluster_i
            
            color = cm.nipy_spectral(float(i) / n_cluster)
            axs[ind].fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_sil_values, \
                                facecolor=color, edgecolor=color, alpha=0.7)
            axs[ind].text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
            y_lower = y_upper + 10
            
        axs[ind].axvline(x=sil_avg, color="red", linestyle="--")
```

```python
# make_blobs를 통해 clustering을 위한 4개의 클러스터 중심의 500개 2차원 데이터 셋 생성

from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=500, n_features=2, centers=4, cluster_std=1, \
                center_box=(-10.0,10.0), shuffle=True, random_state=1)

# cluster 개수를 2,3,4,5개일때의 클러스터별 실루엣 계수 평균값을 시각화
# 4개의 군집일 때 가장 최적
visualize_silhouette([2,3,4,5], X)
```

## 예시2 -  고객 세그먼테이션
- 다양한 기준으로 고객을 분류하는 기법
- 고객을 분류하는 요소
 - 지역/결혼 여부/성별/소득과 같이 개인의 신상 데이터
 - 고객분류가 상품 판매에 중점을 두고 있기 때문에 더 중요한 분류 요소는 어떤 상품을 얼마나 많은 비용을 써서 얼마나 자주 사용하는가에 기반한 정보로 분류
 - 고객 분류의 주요 목적은 타깃 마케팅이며 고객을 여러 특성에 맞게 세분화해서 그 유형에 따라 맞춤형 마케팅이나 서비스를 제공한다.
 - 평소 많은 돈을 지불해 상품이나 서비스를 이용하고 있다면 VIP 전용 프로그램을 경험했을 것이고 새로운 상품이나 서비스를 적극 이용해 왔다면 프로모션 상품이 출시될 때마다 권유를 받았을 것이다.
- 고객 분류는 고객의 어떤 요소를 기반으로 군집화할 것인가를 결정하는 것이 중요하며 예를들어 Recency(R), Frequency(F), Monetary Value(M)을 의미하는 RFM 기법이 있다.
 - Recency : 가장 최근 상품 구입일에서 오늘까지의 기간
 - Frequency : 상품 구매 횟수
 - Monetary Value : 총 구매 금액

```python
import pandas as pd
import datetime
import math
import numpy as np
import matplotlib.pyplot as plt

retail_df = pd.read_excel('dataset/Online_Retail.xlsx',engine='openpyxl')
retail_df.head()
```

### K-means와 데이터 분포

K-평균 군집 방법은 데이터의 분포 특성을 고려하여 군집화를 수행하는 방법입니다. 일반적으로 K-평균 군집화는 데이터가 원형의 클러스터를 형성하는 경우에 잘 동작합니다. 그러나 데이터가 다른 형태의 분포를 가지거나 왜곡되어 있는 경우에는 K-평균 군집화가 제대로 작동하지 않을 수 있습니다. 이를 보완하기 위해 데이터 분포에 따라 다양한 방법을 사용할 수 있습니다.

1. 데이터 스케일링:
   데이터의 값 범위가 다른 경우에는 스케일링을 통해 데이터를 일정한 범위로 조정합니다. 일반적으로 StandardScaler 또는 MinMaxScaler를 사용하여 데이터를 표준화하거나 정규화합니다. 이를 통해 데이터의 분포를 조정하여 군집화 결과를 개선할 수 있습니다.

2. 로그 변환:
   데이터가 왜곡되거나 비대칭적인 분포를 가질 때에는 로그 변환을 적용할 수 있습니다. 로그 변환은 데이터의 값에 로그 함수를 적용하여 데이터의 스케일을 조정합니다. 이를 통해 데이터의 분포를 정규분포에 가깝게 만들어 군집화 결과를 개선할 수 있습니다.

3. 밀도 기반 군집화:
   K-평균 군집화는 클러스터의 중심과 거리를 기준으로 군집을 형성합니다. 하지만 데이터가 밀도 기반으로 군집을 형성하는 경우에는 DBSCAN과 같은 밀도 기반 군집화 알고리즘을 사용하는 것이 더 적합할 수 있습니다. DBSCAN은 데이터의 밀도에 따라 군집을 형성하기 때문에 데이터 분포에 민감하게 군집화를 수행할 수 있습니다.

이러한 방법들을 사용하여 데이터 분포에 따라 적절한 전처리를 수행하고 K-평균 군집화를 적용하면 데이터의 분포 특성을 잘 반영한 군집화 결과를 얻을 수 있습니다. 데이터 분포에 따라 적절한 전처리 방법을 선택하여 군집화를 수행하는 것이 중요합니다.

```python
def visualize_silhouette(cluster_lists, X_features): 
    
    from sklearn.datasets import make_blobs
    from sklearn.cluster import KMeans
    from sklearn.metrics import silhouette_samples, silhouette_score

    import matplotlib.pyplot as plt
    import matplotlib.cm as cm
    import math
    
    # 입력값으로 클러스터링 갯수들을 리스트로 받아서, 각 갯수별로 클러스터링을 적용하고 실루엣 개수를 구함
    n_cols = len(cluster_lists)
    
    # plt.subplots()으로 리스트에 기재된 클러스터링 만큼의 sub figures를 가지는 axs 생성 
    fig, axs = plt.subplots(figsize=(4*n_cols, 4), nrows=1, ncols=n_cols)
    
    # 리스트에 기재된 클러스터링 갯수들을 차례로 iteration 수행하면서 실루엣 개수 시각화
    for ind, n_cluster in enumerate(cluster_lists):
        
        # KMeans 클러스터링 수행하고, 실루엣 스코어와 개별 데이터의 실루엣 값 계산. 
        clusterer = KMeans(n_clusters = n_cluster, max_iter=500, random_state=0)
        cluster_labels = clusterer.fit_predict(X_features)
        
        sil_avg = silhouette_score(X_features, cluster_labels)
        sil_values = silhouette_samples(X_features, cluster_labels)
        
        y_lower = 10
        axs[ind].set_title('Number of Cluster : '+ str(n_cluster)+'\n' \
                          'Silhouette Score :' + str(round(sil_avg,3)) )
        axs[ind].set_xlabel("The silhouette coefficient values")
        axs[ind].set_ylabel("Cluster label")
        axs[ind].set_xlim([-0.1, 1])
        axs[ind].set_ylim([0, len(X_features) + (n_cluster + 1) * 10])
        axs[ind].set_yticks([])  # Clear the yaxis labels / ticks
        axs[ind].set_xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
        
        # 클러스터링 갯수별로 fill_betweenx( )형태의 막대 그래프 표현. 
        for i in range(n_cluster):
            ith_cluster_sil_values = sil_values[cluster_labels==i]
            ith_cluster_sil_values.sort()
            
            size_cluster_i = ith_cluster_sil_values.shape[0]
            y_upper = y_lower + size_cluster_i
            
            color = cm.nipy_spectral(float(i) / n_cluster)
            axs[ind].fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_sil_values, \
                                facecolor=color, edgecolor=color, alpha=0.7)
            axs[ind].text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
            y_lower = y_upper + 10
            
        axs[ind].axvline(x=sil_avg, color="red", linestyle="--")
```

```python
### 여러개의 클러스터링 갯수를 List로 입력 받아 각각의 클러스터링 결과를 시각화 
def visualize_kmeans_plot_multi(cluster_lists, X_features):
    
    from sklearn.cluster import KMeans
    from sklearn.decomposition import PCA
    import pandas as pd
    import numpy as np
    
    # plt.subplots()으로 리스트에 기재된 클러스터링 만큼의 sub figures를 가지는 axs 생성 
    n_cols = len(cluster_lists)
    fig, axs = plt.subplots(figsize=(4*n_cols, 4), nrows=1, ncols=n_cols)
    
    # 입력 데이터의 FEATURE가 여러개일 경우 2차원 데이터 시각화가 어려우므로 PCA 변환하여 2차원 시각화
    pca = PCA(n_components=2)
    pca_transformed = pca.fit_transform(X_features)
    dataframe = pd.DataFrame(pca_transformed, columns=['PCA1','PCA2'])
    
     # 리스트에 기재된 클러스터링 갯수들을 차례로 iteration 수행하면서 KMeans 클러스터링 수행하고 시각화
    for ind, n_cluster in enumerate(cluster_lists):
        
        # KMeans 클러스터링으로 클러스터링 결과를 dataframe에 저장. 
        clusterer = KMeans(n_clusters = n_cluster, max_iter=500, random_state=0)
        cluster_labels = clusterer.fit_predict(pca_transformed)
        dataframe['cluster']=cluster_labels
        
        unique_labels = np.unique(clusterer.labels_)
        markers=['o', 's', '^', 'x', '*']
       
        # 클러스터링 결과값 별로 scatter plot 으로 시각화
        for label in unique_labels:
            label_df = dataframe[dataframe['cluster']==label]
            if label == -1:
                cluster_legend = 'Noise'
            else :
                cluster_legend = 'Cluster '+str(label)           
            axs[ind].scatter(x=label_df['PCA1'], y=label_df['PCA2'], s=70,\
                        edgecolor='k', marker=markers[label], label=cluster_legend)

        axs[ind].set_title('Number of Cluster : '+ str(n_cluster))    
        axs[ind].legend(loc='upper right')
    
    plt.show()
```

```python
visualize_silhouette([2,3,4,5], X_features_scaled)
visualize_kmeans_plot_multi([2,3,4,5],X_features_scaled)
```
