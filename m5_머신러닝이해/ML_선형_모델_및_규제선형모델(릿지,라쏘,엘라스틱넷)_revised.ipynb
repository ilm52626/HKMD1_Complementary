여러분은 회사에서 새로운 제품을 출시하려고 합니다. 제품의 판매량을 예측하기 위해 선형모델을 사용하려고 합니다.

여러분은 판매량과 광고 비용, 경쟁 업체의 가격 등과 같은 정보들을 수집했어요. 이 정보들을 사용하여 선형모델을 학습시켜서 어떤 요소들이 제품의 판매량에 영향을 주는지 알아습니다.

이제 새로운 제품을 출시하고, 광고 비용과 경쟁 업체의 가격을 알고 있다면, 선형모델을 사용해서 제품의 예상 판매량을 예측할 수 있습니다.

예를 들어, 광고 비용이 1000달러이고 경쟁 업체의 가격이 50달러라는 정보를 가지고 있다면, 선형모델을 사용해서 제품의 예상 판매량을 알아낼 수 있습니다.

이렇게 선형모델은 여러분에게 새로운 제품의 성공 가능성을 알려주고, 비즈니스 결정을 내리는 데 도움을 줄 수 있습니다.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# 판매량, 광고 비용, 경쟁 업체의 가격 데이터
sales = np.array([100, 200, 300, 400, 500])  # 판매량
ad_costs = np.array([[1000], [2000], [3000], [4000], [5000]])  # 광고 비용
competitor_prices = np.array([[50], [40], [30], [20], [10]])  # 경쟁 업체의 가격

# 선형 회귀 모델 생성
model = LinearRegression()
model.fit(np.concatenate((ad_costs, competitor_prices), axis=1), sales)

# 새로운 정보에 대한 예측
new_ad_cost = np.array([[1500]])  # 새로운 광고 비용
new_competitor_price = np.array([[25]])  # 새로운 경쟁 업체의 가격
predicted_sales = model.predict(np.concatenate((new_ad_cost, new_competitor_price), axis=1))


# 새로운 광고 비용과 경쟁 업체의 가격 정보를 입력하여 알아낸 제품의 예상 판매량
print(predicted_sales)
```

## 특징

선형모델(Linear Model)은 입력 변수들의 선형 조합을 통해 출력 변수를 예측하는 머신러닝 모델입니다. 선형모델은 입력 변수들 간의 선형 관계를 가정하고, 이를 기반으로 예측을 수행합니다.

선형모델은 주어진 데이터에 대해 최적의 선을 찾는 것으로 특징이 있습니다. 이 선은 입력 변수들의 가중치와 편향(bias)으로 정의되며, 입력 변수들과 가중치의 곱을 모두 더한 후 편향을 더함으로써 예측 값을 계산합니다. 이렇게 선형모델은 입력 변수들과 가중치의 선형 조합으로 출력 값을 예측하는 형태를 갖습니다.

선형모델은 간단하고 해석하기 쉬우며, 다양한 분야에서 사용됩니다. 예를 들어, 회귀 문제에서는 연속적인 출력 변수를 예측하는 데 사용되고, 분류 문제에서는 이진 분류나 다중 클래스 분류를 수행하는 데 사용됩니다.

선형모델은 다양한 변형과 확장이 가능하며, 각각은 다른 목적과 제약 조건에 적합합니다. 일반적인 선형모델로는 선형 회귀(Linear Regression)와 로지스틱 회귀(Logistic Regression)가 있으며, 이를 기반으로 한 다른 모델들도 널리 사용됩니다.

선형모델은 입력 변수들 간의 선형 관계를 가정하기 때문에, 입력 변수들이 비선형적인 관계를 가질 경우 성능이 제한될 수 있습니다. 이를 극복하기 위해 선형모델에 다항식이나 상호작용 항을 추가하는 등의 확장을 시도할 수 있습니다. 또한, 정규화(regularization) 방법을 사용하여 과적합을 방지하고 일반화 성능을 개선할 수도 있습니다.



##다중공선성(Multicollinearity)


여러 가게들이 모여 있는 작은 상가는 주민들에게 다양한 상품과 서비스를 제공하고 있습니다. 각 가게는 자신만의 독특한 특성과 장점을 가지고 있었고, 서로 경쟁하며 고객들의 마음을 사로잡으려고 노력했습니다.

그러나 어느 날 상가 주변에 새로운 가게가 열렸습니다. 이 가게는 다른 가게들과는 달리 모든 상품과 서비스를 한 자리에서 제공하는 것이었습니다. 이 가게는 다른 가게들에게 큰 경쟁 압력을 주었고, 상가 내의 가게들은 매출이 감소하고 어려움을 겪게 되었습니다.

이 문제를 해결하기 위해 상가 주인들은 다중공선성에 대해 고민하기 시작했습니다. 다중공선성은 여러 가게들이 유사한 상품과 서비스를 제공하여 고객들의 선택이 중복되고 경쟁이 치열해지는 현상입니다.

그들은 다양성과 특색을 강조하여 각 가게들이 독특한 분야에 집중하도록 결정했습니다. 예를 들어, 한 가게는 커피와 디저트에 특화되어 다른 가게들과 차별화를 꾀하고, 다른 가게는 신선한 농산물과 유기농 제품을 판매하여 다른 시장을 겨냥했습니다.

이렇게 각 가게들이 다중공선성을 고려하여 자신만의 독특한 분야에 집중하고 차별화된 경험을 제공하자, 상가는 다시 번영하게 되었습니다. 고객들은 다양한 선택지를 가지고 있으면서도 각 가게의 특색과 장점을 살려 고민 없이 원하는 상품과 서비스를 선택할 수 있게 되었습니다.


## 특징

다중공선성(Multicollinearity)은 회귀 분석에서 독립 변수들 간에 강한 선형 관계가 존재하는 상황을 의미합니다. 즉, 한 독립 변수가 다른 독립 변수들과 상관 관계가 높아서 이들 변수들이 서로 설명력을 공유하거나 중복된 정보를 포함하는 경우를 말합니다.

다중공선성은 회귀 분석에서 중요한 문제로 여겨집니다. 이는 다음과 같은 이유로 인해 문제가 될 수 있습니다:
1. 계수 추정의 불안정성: 독립 변수들 간에 강한 선형 관계가 존재하면 회귀 분석에서 계수의 추정이 불안정해지고 변동성이 증가합니다.
2. 계수의 해석의 어려움: 독립 변수들 간에 강한 상관 관계가 있으면 어떤 변수가 종속 변수와 실제로 얼마나 관련이 있는지를 파악하기 어려워집니다.
3. 변수 선택의 어려움: 다중공선성이 존재하면 어떤 독립 변수를 모델에 포함시킬지 선택하는 것이 어려워집니다.

다중공선성을 해결하기 위해 몇 가지 방법을 사용할 수 있습니다. 주요한 방법은 다음과 같습니다:
1. 변수 선택: 상관 관계가 강한 독립 변수들 중에서 하나를 선택하여 제거하거나, 변수들을 합쳐서 새로운 변수를 생성하는 등의 변수 선택 방법을 사용
2. 차원 축소: PCA등 고차원 데이터를 저차원으로 변환하는 기법을 사용하여 변수의 갯수를 줄여 다중공선성을 감소
3. 변수 변환: 로그 변환, 차분 변환 등의 변수 변환 기법을 사용하여 변수들을 변환하여 다중공선성 감소
4. 규제 기법: 릿지 회귀, 라쏘 회귀, 엘라스틱넷 등의 규제 기법을 사용하여 변수들의 계수를 제한함으로써 다중공선성을 감소

이러한 방법들을 적용하여 다중공선성을 관리하면 회귀 분석의 안정성을 높이고, 모델의 신뢰도와 해석력을 개선할 수 있습니다.


## 회귀 평가지표

회귀 평가지표는 회귀 모델의 성능을 측정하고 평가하는 데 사용되는 지표들을 말합니다. 회귀 모델은 입력 변수와 출력 변수 간의 관계를 모델링하는데 사용되며, 회귀 평가지표는 모델이 얼마나 잘 예측하는지를 평가하는 데 도움을 줍니다. 일반적으로 회귀 모델의 예측값과 실제 값 사이의 차이를 계산하여 모델의 정확성을 평가합니다.

가장 일반적으로 사용되는 회귀 평가지표는 다음과 같습니다:

1. 평균 제곱 오차(Mean Squared Error, MSE): 실제 값과 예측 값 사이의 차이의 제곱을 평균한 값입니다. 예측 오차의 크기를 나타내며, 값이 작을수록 모델의 예측이 정확합니다.

2. 평균 절대 오차(Mean Absolute Error, MAE): 실제 값과 예측 값 사이의 차이의 절댓값을 평균한 값입니다. 예측 오차의 크기를 나타내며, 값이 작을수록 모델의 예측이 정확합니다.

3. 결정 계수(Coefficient of Determination, R^2): 예측 값이 종속 변수의 변동성을 얼마나 설명하는지를 나타내는 지표입니다. 0부터 1까지의 값을 가지며, 값이 1에 가까울수록 모델의 예측이 정확합니다.

4. 수정된 결정 계수(Adjusted R^2): R^2 값을 독립 변수의 개수와 샘플 크기에 맞게 수정한 지표입니다. 다중 회귀 분석에서 독립 변수의 개수가 증가할수록 R^2 값이 증가하는 경향이 있기 때문에, 이를 보정하여 모델의 설명력을 평가합니다.

이 외에도 회귀 평가지표로는 평균 제곱근 오차(Root Mean Squared Error, RMSE), 설명된 분산(Explained Variance) 등이 있습니다. 이러한 평가지표들을 통해 회귀 모델의 성능을 측정하고 다른 모델들과 비교할 수 있습니다.


### 음수값 평가지표

Scikit-learn 라이브러리의 Scoring 함수는 일반적으로 높은 값이 좋은 모델 성능을 의미하는 지표를 사용합니다. 그러나 'neg_mean_absolute_error'와 같이 일부 지표는 낮은 값이 더 좋은 모델 성능을 의미하기 때문에 음수 값을 반환합니다.

'neg_mean_absolute_error'는 평균 절대 오차(MAE)의 음수 값입니다. MAE는 예측값과 실제값 사이의 차이의 절댓값을 평균한 값으로, 작을수록 모델의 예측이 정확합니다. Scikit-learn은 일관성을 위해 모든 Scoring 함수에서 높은 값이 좋은 성능을 의미하도록 정의하고 있습니다. 따라서 MAE의 반대 개념인 음수 MAE를 반환하여, 높은 값일수록 모델의 성능이 좋음을 나타냅니다.

이렇게 Scoring 함수가 음수 값을 반환하는 이유는 일관성을 유지하고, 모든 지표가 높은 값이 좋은 성능을 의미하도록 통일하기 위함입니다. 따라서 음수 MAE를 사용하여 모델의 성능을 평가하고 비교할 수 있습니다.

## 규제 선형모델

규제 선형 모델(Regularized Linear Model)은 선형 모델에 정규화(regularization)를 적용한 모델을 의미합니다. 정규화는 모델의 복잡도를 제어하여 과적합을 방지하고 일반화 성능을 향상시키는 방법입니다.

일반적인 선형 모델은 주어진 입력 변수들 간의 선형 관계를 학습하고, 이를 통해 출력 변수를 예측합니다. 하지만 입력 변수들 간의 관계가 복잡하거나 고차원 데이터에서는 모델이 과적합되는 경향이 있습니다. 이를 해결하기 위해 규제 선형 모델은 추가적인 제약 조건을 도입하여 모델의 복잡도를 제한합니다.

가장 널리 사용되는 규제 선형 모델은 릿지 회귀(Ridge Regression)와 라쏘 회귀(Lasso Regression), 엘라스틱넷(ElastcitNet)입니다. 이들은 모델의 손실 함수에 정규화 항을 추가하여 모델의 가중치를 제한합니다.

1.	릿지 회귀 (Ridge Regression):


  - L2 규제를 적용하여 가중치의 크기를 제한하는 방식입니다.

  -손실 함수에 L2 노름(norm)을 추가하여 가중치의 제곱에 비례하는 페널티를 부여합니다.

  -가중치를 가능한 작게 유지하면서 모든 입력 변수를 고려하는 효과가 있습니다.

  -가중치 감소(Weight Decay)라고도 불립니다.

2.	라쏘 회귀 (Lasso Regression):


  - L1 규제를 적용하여 가중치의 절댓값을 제한하는 방식입니다.

  -손실 함수에 L1 노름을 추가하여 가중치의 절댓값에 비례하는 페널티를 부여합니다.

  -가중치를 0으로 만들어 특정 변수를 선택하거나 희소성(sparcity)을 가진 모델을 생성할 수 있습니다.

  -변수 선택(Variable Selection)이 가능한 모델입니다.

3.	엘라스틱넷 (ElasticNet):


  - L1 규제와 L2 규제를 함께 적용하는 모델입니다.
  - 	L1 규제는 변수 선택과 희소성을 유지하는 데 도움을 주고, L2 규제는 다중공선성(multicollinearity) 문제를 완화하는 데 도움을 줍니다.
  - L1 규제와 L2 규제의 비율을 조절하는 매개변수로 모델을 조정할 수 있습니다.


이러한 규제 선형 모델들은 과적합을 방지하고 일반화 성능을 향상시키는 장점이 있습니다. 또한, 변수 선택이나 특성 중요도 평가 등의 추가적인 기능을 제공할 수도 있습니다.



```python
# 릿지 회귀
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score

ridge = Ridge(alpha=10)
neg_mse_scores = cross_val_score(ridge, X_data, y_target,\
                                scoring="neg_mean_squared_error"\
                                ,cv=5)
rmse_scores = np.sqrt(-1*neg_mse_scores)
avg_rmse = np.mean(rmse_scores)
print('5 folds의 개별 Negative MSE scores :', np.round(neg_mse_scores,3))
print('5 folds의 개별 RMSE scores :', np.round(rmse_scores,3))
print('5 folds의 평균 RMSE scores : {0:.3f}'.format(avg_rmse))
```

```python
alphas = [0, 0.1, 1, 10, 100]
for alpha in alphas:
    ridge = Ridge(alpha=alpha)
    
    neg_mse_scores = cross_val_score(ridge, X_data, y_target,\
                                    scoring="neg_mean_squared_error"\
                                    ,cv=5)
    avg_rmse = np.mean(np.sqrt(-1*neg_mse_scores))
    print('alpha {}일 때 5 folds의 평균 RMSE : {:.3f}'.format(alpha,\
                                                         avg_rmse))
```

```python
# 라쏘 회귀
from sklearn.linear_model import Lasso, ElasticNet
def get_linear_reg_eval(model_name,params=None,X_data_n=None, 
                        y_target_n=None, verbose=True):
    coeff_df = pd.DataFrame()
    if verbose : print('####### ', model_name , '#######')
    for param in params:
        if model_name == 'Ridge':model = Ridge(alpha=param)
        elif model_name == 'Lasso':model = Lasso(alpha=param)
        elif model_name == 'ElasticNet':model = ElasticNet(alpha=param, \
                                                          l1_ratio=0.7)
        neg_mse_scores = \
        cross_val_score(model,X_data_n,y_target_n, \
                        scoring="neg_mean_squared_error", cv=5)
        avg_rmse = np.mean(np.sqrt(-1*neg_mse_scores))
        print('alpha {}일 때 5 폴드 세트의 평균 RMSE: {:3f}'\
              .format(param, avg_rmse))
        model.fit(X_data,y_target)
        coeff = pd.Series(data=model.coef_, index=X_data.columns)
        colname = 'alpha:'+str(param)
        coeff_df[colname] = coeff
    return coeff_df    
```

```python
lasso_alphas = [0.07,0.1,0.5,1,3]
coeff_lasso_df = get_linear_reg_eval('Lasso',params=lasso_alphas,\
                                    X_data_n=X_data,y_target_n=y_target)
coeff_lasso_df
```

```python
# 선형 회귀 모델을 위한 데이터 변환
# X_data 변환
from sklearn.preprocessing import StandardScaler,MinMaxScaler,\
PolynomialFeatures
def get_scaled_data(method='None',p_degree=None, input_data=None):
    if method=='Standard':
        scaled_data = StandardScaler().fit_transform(input_data)
    elif method == 'MinMax':
        scaled_data = MinMaxScaler().fit_transform(input_data)
    elif method == 'Log':
        scaled_data = np.log1p(input_data)
    else:
        scaled_data = input_data
    if p_degree != None:
        scaled_data = PolynomialFeatures(degree=p_degree,include_bias=False).\
        fit_transform(scaled_data)
    return scaled_data        
```
