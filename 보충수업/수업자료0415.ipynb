# 로그의 의미, 로그함수란

- 로그란 어떤 수를 다른 수로 나눈 비율을 나타내는 수학적 개념입니다. 일반적으로 10이나 자연로그(e)를 밑으로 사용합니다. 예를 들어, 100을 10으로 나눈 값은 10이므로, 로그 함수의 값은 2가 됩니다. 또 다른 예로, 1,000을 10으로 나눈 값은 100이므로, 로그 함수의 값은 3이 됩니다. 로그함수는 데이터 분석에서 다양한 용도로 활용됩니다. 로그함수의 값은 보통 양수를 입력으로 받으며, 출력은 실수값이 됩니다. 로그 함수를 적용하여 데이터의 분포를 변환하면, 정규 분포에 가까운 형태를 얻을 수 있습니다. 이러한 변환은 분석 결과를 더욱 해석하기 쉽게 만들어주며, 일반적으로 데이터 분석에서 매우 유용합니다.

#밀도함수와 확률밀도함수

- 밀도는 단위 면적당 어떤 현상이 발생하는 빈도를 의미합니다. 확률밀도함수는 밀도와 비슷한 개념으로, 연속형 변수의 값을 갖는 확률변수의 밀도를 나타냅니다. 이것은 분포 함수의 미분으로 구할 수 있으며, 어떤 구간에서 변수가 발생할 확률은 그 구간에서 확률밀도함수의 면적과 같습니다. cdf와 ppf는 확률밀도함수와 관련이 있습니다. cdf는 누적분포함수를 의미하며, ppf는 누적분포함수의 역함수로, cdf가 주어졌을 때 어떤 값이 나올 확률을 계산하는 데 사용됩니다.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# 타이타닉 데이터에서 승객의 나이에 대한 밀도 함수와 히스토그램 출력
sns.kdeplot(titanic['age'], shade=True)
plt.hist(titanic['age'], bins=20, alpha=0.5)
plt.title('Density Function and Histogram of Age')
plt.xlabel('age')
plt.ylabel('Density')
plt.show()
```

cdf와 ppf는 확률분포를 나타내는 함수 중 일부입니다. 이 함수들은 주로 통계학에서 사용되며, 확률을 계산하거나, 분포를 분석하는 데에 사용됩니다.

- cdf는 누적분포함수를 의미합니다. 누적분포함수는 특정 값 이하의 확률을 나타내는 함수입니다. 예를 들어, 주사위를 던져서 나오는 눈의 수에 대한 누적분포함수를 생각해보면, 누적분포함수는 주사위를 던졌을 때 나온 눈의 수가 1이거나 2이거나 3이거나 4이거나 5이거나 6일 때의 눈의 수가 이하인 확률을 더한 것을 나타냅니다.

- ppf는 누적분포함수의 역함수를 의미합니다. 역함수는 입력값과 출력값이 서로 바뀐 함수입니다. 즉, 누적분포함수의 출력값을 입력으로 받아서, 해당 출력값이 나오는 입력값을 계산하는 함수입니다. 예를 들어, 주사위를 던져서 나오는 눈의 수에 대한 누적분포함수의 역함수를 생각해보면, 이것은 특정한 확률이 주어졌을 때, 해당 확률을 넘는 눈의 수를 계산하는 함수입니다.

따라서, cdf와 ppf는 확률분포를 분석하는 데에 사용되는 함수이며, 이 함수들은 입력값과 출력값이 서로 다르기 때문에, 그래프로 시각화하기가 어렵습니다. 하지만, 이러한 함수를 사용하면, 확률분포를 더욱 정확하게 분석할 수 있습니다.

```python
from scipy.stats import norm


titanic.dropna(subset=['age'], inplace=True)

# 나이 변수의 평균과 표준편차 계산
age_mean = np.mean(titanic['age'])
age_std = np.std(titanic['age'])

# 나이 변수를 표준정규분포로 변환
age_z = (titanic['age'] - age_mean) / age_std

# 나이 변수에 대한 cdf 계산
age_cdf = norm.cdf(age_z)

# cdf 그래프 출력
plt.plot(titanic['age'], age_cdf)
plt.title('CDF of Age')
plt.xlabel('Age')
plt.ylabel('Cumulative Probability')
plt.show()

# 나이 변수에 대한 ppf 계산
age_ppf = norm.ppf(age_cdf)

# ppf 그래프 출력
plt.plot(age_cdf, age_ppf)
plt.title('PPF of Age')
plt.xlabel('Cumulative Probability')
plt.ylabel('Age')
plt.show()
```
