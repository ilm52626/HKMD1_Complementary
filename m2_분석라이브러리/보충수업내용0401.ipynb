2. 직교행렬에 대한 추가적인 설명


image.png

직교행렬은 수학적인 개념으로 이해하기가 어렵습니다. 그러나 직교행렬의 기하학적인 의미를 이해한다면 직관적으로 이해할 수 있습니다.

기하학적으로, 직교행렬은 선형 대수학에서 벡터를 회전시키는데 사용됩니다. 예를 들어, 2차원 공간에서 (1, 0) 벡터와 (0, 1) 벡터는 서로 직교하며, (1, 0) 벡터를 반시계 방향으로 90도 회전하면 (0, 1) 벡터가 됩니다.

이와 같이, 직교행렬은 벡터를 회전시키는데 사용됩니다. 직교행렬의 열벡터는 서로 직교하며 크기가 1인 벡터이기 때문에, 이를 사용하여 벡터를 회전시킬 때 벡터의 크기는 변하지 않고 방향만 바뀌게 됩니다.

예를 들어, 2차원 공간에서 (1, 0) 벡터와 (0, 1) 벡터를 열로 가지는 직교행렬은 다음과 같습니다.

```python
[[1, 0],
 [0, 1]]
```

이 행렬은 아무런 회전도 하지 않기 때문에, (1, 0) 벡터를 이 행렬과 곱하면 결과는 (1, 0) 벡터가 됩니다. 반면, 다음과 같은 직교행렬은 90도 반시계 방향으로 회전하는 행렬입니다.

```python
[[0, -1],
 [1,  0]]
```

이 행렬과 (1, 0) 벡터를 곱하면 결과는 (0, 1) 벡터가 됩니다. 즉, 직교행렬을 사용하여 벡터를 회전시키면 벡터의 방향이 바뀌지만 크기는 변하지 않는 것을 알 수 있습니다.

따라서, 직교행렬은 기하학적으로 벡터를 회전시키는데 사용되며, 수학적인 의미보다는 이러한 기하학적인 의미를 이해하는 것이 더 쉬울 수 있습니다.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3차원 직교행렬 생성
A = np.array([[1, 0, 0], [0, -1, 0], [0, 0, 1]])
A
```

```python
# 각 열 벡터 그리기
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
origin = [0], [0], [0]
ax.quiver(*origin, A[:, 0], A[:, 1], A[:, 2], color=['r','b','g'], arrow_length_ratio=0.1)
ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([-1.5, 1.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Orthogonal Matrix')
plt.show()
```

3. merge()에 대한 전체적인 설명(outer, inner, left, right), + join(suffix...)설명

```python
import pandas as pd

df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'], 'value': [5, 6, 7, 8]})
```

image.png

```python
df1
```

```python
df2
```

```python
# inner join
inner_join = pd.merge(df1, df2, on='key', how='inner')
inner_join
```

```python
# left join
left_join = pd.merge(df1, df2, on='key', how='left')
left_join
```

```python
# right join
right_join = pd.merge(df1, df2, on='key', how='right')
right_join

```

```python
# outer join
outer_join = pd.merge(df1, df2, on='key', how='outer')
outer_join
```
