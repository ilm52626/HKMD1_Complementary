##확률밀도함수

한 가게에서는 하루마다 방문 고객 수를 기록하고 있습니다. 가게 주인은 이 데이터를 활용하여 효율적인 운영 방법을 찾고자 합니다.

그러나 매일매일 수십명씩 방문한 고객들의 수를 하나하나 세어 기록하는 일은 어렵습니다. 주인은 하루 동안 방문한 고객 수를 총합하여 하나의 숫자로만 기록하기로 하였습니다.

그렇게 기록된 데이터를 가지고 어떤 날은 많은 고객들이 방문했고, 또 다른 날은 적은 고객들이 방문했다는 사실을 알 수 있습니다. 이러한 데이터를 효율적으로 분석하기 위해서는, 이 데이터가 어떤 분포를 따르는지 알아내야 합니다.

**확률밀도함수**를 이용하여 해당 데이터를 묘사할 수 있습니다. 이 때, 확률밀도함수는 해당 데이터가 따르는 확률분포를 나타내는 함수 중 하나로, 어떤 사건이 일어날 확률을 그래프로 나타내는 방법입니다.

확률밀도함수 그래프는 가로축에 가능한 모든 사건을, 세로축에는 그 사건이 일어날 확률을 나타내는 함수 값으로 표시됩니다. 이를 통해, 어떤 날은 고객이 적게 방문했을 때 그래프의 면적이 작아지고, 반대로 고객이 많이 방문했을 때 그래프의 면적이 커지는 것을 볼 수 있습니다.

확률밀도함수는 이러한 분석을 효율적으로 할 수 있게 해주며, 데이터 분석에서 매우 중요한 개념입니다. 파이썬으로 확률밀도함수를 계산하는 것도 매우 간단하며, 분석에 필요한 다양한 정보를 추출할 수 있습니다.

```python
# 평균이 10, 표준편차가 3인 정규분포의 확률밀도함수를 그래프로 표현
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
x_plot = np.arange(0,20,0.1)
plt.plot(x_plot, stats.norm.pdf(x=x_plot,loc=10,scale=3), color='black')
plt.show()
```

##누적분포함수

우리는 지도를 보면서 목적지까지 가는 것이 얼마나 멀리 남았는지, 얼마나 많은 거리를 이미 지나왔는지 알 수 있습니다. 마찬가지로 확률분포에서도, 어떤 사건이 일어날 확률을 보여주는 누적분포함수를 통해 우리가 얼마나 많은 확률을 이미 지나갔는지, 얼마나 많은 확률이 남아있는지 알 수 있습니다.

주사위를 던져서 나오는 수의 확률분포를 생각해보면, 누적분포함수는 주사위를 던져서 나온 수가 특정 숫자보다 작거나 같을 확률을 보여줍니다. 즉, 1부터 6까지의 모든 수에 대해 누적분포함수를 계산하면, 1, 2, 3, 4, 5, 6이 나올 확률의 누적분포를 알 수 있습니다.

누적분포함수는 이전 확률값을 더해 나가면서 계산되기 때문에, 누적분포함수의 값은 점점 커집니다. 즉, 어떤 사건이 일어날 확률이 작을수록 누적분포함수의 값은 작아지고, 확률이 커질수록 누적분포함수의 값은 커집니다.

확률분포를 이해하고 분석하는데 누적분포함수는 매우 유용합니다. 예를 들어, 누적분포함수를 사용하여 어떤 사건이 발생할 확률을 추정할 수 있습니다. 이를 통해 데이터 분석에서 예측 모델링이나 시뮬레이션 등에 활용할 수 있습니다.

위 내용을 코드로 표현하면 다음과 같습니다.

```python
import numpy as np
import matplotlib.pyplot as plt

# 확률분포 함수 정의
def probability_distribution(x):
    return x**2/10

# 누적분포 함수 정의
def cumulative_distribution(x):
    cdf = []
    for i in range(len(x)):
        cdf.append(np.trapz(probability_distribution(x[:i+1]), x[:i+1]))
    return cdf

# x값 생성
x = np.linspace(-10, 10, 1000)

# 누적분포 함수 시각화
plt.plot(x, cumulative_distribution(x), label='Cumulative Distribution Function')
plt.xlabel('X')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function of Probability Distribution')
plt.legend()
plt.show()
```

```python
# 평균이 1, 표준편차가 2인 정규분포의 누적분포함수를 그래프로 표현

import matplotlib.pyplot as plt
x_plot = np.arange(-3,8,0.1)
plt.plot(x_plot, stats.norm.cdf(x=x_plot,loc=1,scale=2), color='black')
```

##역함수

한 마을에 놀이공원이 있다고 상상해봅시다. 그리고 이 놀이공원에는 자전거 대여소가 있습니다. 자전거 대여소에서는 1시간에 5,000원씩 자전거를 빌릴 수 있습니다.

그러나 자전거를 빌리려면 우리는 엄마나 아빠에게 돈을 받아야 합니다. 엄마가 20,000원을 우리에게 주었다고 가정해봅시다. 그러면 이제 우리는 얼마나 오랫동안 자전거를 빌릴 수 있을까요?

역함수를 사용하면 쉽게 자전거 대여 시간을 계산할 수 있습니다. 역함수는 입력과 출력을 반대로 바꾸는 함수인데, 여기에서는 우리가 가진 돈을 입력값으로, 빌릴 수 있는 자전거 대여 시간을 출력값으로 생각합니다. 그러면 역함수를 사용하면, 우리가 얼마나 오랫동안 자전거를 빌릴 수 있는지 쉽게 계산할 수 있습니다.

자전거 대여소에서는 1시간에 5,000원을 받습니다. 따라서 역함수는 y = x/5,000이 됩니다. 이때, 우리가 가진 돈이 20,000원이라면, 역함수에 x = 20,000을 넣어서 y값을 계산하면 됩니다. 즉, y = 20,000 / 5,000 = 4시간이므로, 우리는 4시간 동안 자전거를 빌릴 수 있습니다.

이처럼 역함수는 입력과 출력을 반대로 바꾸는 함수로, 여러 상황에서 유용하게 활용됩니다. 예를 들어, 우리가 가진 돈과 물건의 가격을 알고 있다면, 역함수를 사용하여 물건을 얼마까지 살 수 있는지 계산할 수 있습니다. 또한 기업이나 투자자들은 역함수를 사용하여 예측 모델링이나 시뮬레이션 등에 활용할 수 있습니다.

```python
def inverse_function(x):
    return x / 5000
    
money = 20000
time = inverse_function(money)
print(time)


import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 30001, 1000)
y = inverse_function(x)

plt.plot(x, y)
plt.xlabel("Money")
plt.ylabel("Bike Rental Time")
plt.title("Inverse Function")
plt.show()
```

###역함수의 그래프
역함수의 그래프는 어떤 함수의 입력과 출력을 반대로 바꾼 함수를 그래프로 나타낸 것입니다. 함수 f(x) = x^2 의 그래프는 y축과 x축이 교차하는 원점을 중심으로 오른쪽으로 볼록한 모양을 갖습니다. 그리고 이 함수의 역함수는 y = √x이므로, 이 역함수의 그래프는 x축과 y축이 교차하는 원점을 중심으로 왼쪽 위로 향하는 모양을 갖습니다.

이렇게 그래프를 이용하면, 역함수의 성질을 시각적으로 파악할 수 있습니다. 예를 들어, 역함수의 그래프가 x축에 접하는 지점이 있다면, 이는 원래 함수에서 특정한 입력값에 대한 출력값이 여러 개 존재하는 것을 의미합니다. 또한, 역함수의 그래프가 좌우 대칭인 경우에는 원래 함수가 홀수 함수인 것을 의미하며, 이는 입력값이 음수일 때와 양수일 때의 결과값이 서로 반대인 함수를 나타냅니다.

그래프를 통해 역함수의 성질을 파악하는 것은 데이터 분석에서도 중요합니다. 한 예로 어떤 기업의 제품 판매량 데이터를 분석하고자 할 때 역함수의 그래프를 이용하여 제품 가격과 판매량 사이의 관계를 파악할 수 있습니다. 이를 통해 기업은 최적의 가격대를 찾아내고 이를 기반으로 경쟁력 있는 가격 정책을 수립할 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

# 함수 정의: y = x^2
def func(x):
    return x ** 2

# 그래프를 그리기 위해 x값 범위 지정
x = np.linspace(-5, 5, 100)
y = func(x)

# 그래프 그리기
plt.plot(x, y)
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# 역함수 정의: y = sqrt(x)
def inverse_func(x):
    return np.sqrt(x)

# 그래프를 그리기 위해 x값 범위 지정
x_inv = np.linspace(0, 25, 100)
y_inv = inverse_func(x_inv)

# 그래프 그리기
plt.plot(x_inv, y_inv)
plt.title("y = sqrt(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

##Relu(Recitifed Linear Unit) 함수

- 인공 신경망에서 가장 많이 사용되는 활성화 함수 중 하나
- 입력값이 0보다 작을 경우 0으로, 0 이상일 경우 입력값을 그대로 출력

ReLU 함수는 마치 걸러내는 기계와 같습니다. 우리는 걸러내고 싶지 않은 것들은 그대로 지나가게 하고, 걸러내고 싶은 것들은 버리는 것입니다. 우리가 바다에서 고기를 잡는다고 해봅시다. 걸러내고 싶지 않은 것들, 즉 바다 속에서 우리가 원하지 않는 돌이나 해초 등은 그대로 지나가게 하고, 걸러내고 싶은 것들, 즉 고기들은 우리가 잡아서 가지고 올 것입니다. ReLU 함수도 이와 비슷한 역할을 합니다. 양수인 값은 그대로 지나가게 하고, 음수인 값은 걸러내는 것입니다.

```python
xx = np.linspace(-10, 10, 100)
plt.figure(figsize=(10,6))
plt.plot(xx, np.maximum(xx, 0))
plt.title("max(x,0) 또는 ReLU")
plt.xlabel("$x$")
plt.ylabel("$ReLU(x)$")
plt.show()
```

##지수함수
- 정해진 어떤 값을 입력값으로 거듭제곱하여 출력을 구하는 함수
- 거듭 제곱을 할 때 아래에 있는 수를 밑이라고 하며 밑을 오일러 수 e(약 2.718)로 하여 거듭제곱하는 함수를 지수함수라고 한다.
- $y=e^x$, y=exp(x) = exp x 로 표시

케이크가 있다고 해봅시다. 이 케이크를 자를 때, 케이크의 높이에 따라 잘린 조각의 크기가 달라집니다. 케이크의 높이가 높을수록 큰 조각이 잘리고, 케이크의 높이가 낮을수록 작은 조각이 잘립니다. 이때 케이크의 높이가 지수함수에서의 밑이 되는 수라고 생각해보면, 케이크의 높이가 높을수록 큰 값을 가지는 함수를 만들 수 있습니다.

예를 들어, 밑이 2인 지수함수 y=2^x 를 그려보면, x 값이 증가할수록 y 값이 기하급수적으로 증가하는 것을 볼 수 있습니다. 이는 마치 케이크의 높이가 높을수록 큰 조각이 잘리는 것과 비슷합니다. 이처럼 지수함수는 큰 수를 나타내는 데에 매우 유용한 함수입니다.

###로지스틱 함수

- 지수함수를 변형한 함수로 시그모이드 함수의 하나

로지스틱 함수는 우리가 선택지 중 하나를 선택할 때 사용하는 함수입니다. 예를 들어, 어떤 사람이 배가 고프다면, 그 사람은 "배가 고프다" 라는 선택지와 "배가 안 고프다" 라는 선택지 중 하나를 선택해야 합니다. 로지스틱 함수는 이와 같은 선택지 중 하나를 선택할 때 사용하는 함수입니다.

로지스틱 함수는 입력값에 따라 출력값이 달라집니다. 입력값이 커지면 출력값이 1에 가까워지고, 입력값이 작아지면 출력값이 0에 가까워집니다. 따라서 입력값이 임계치(threshold)보다 크면 1, 작으면 0을 출력합니다.

수학 시험에서 60점 이상이면 합격, 60점 미만이면 불합격으로 판단하는 경우를 생각해봅시다. 이때, 로지스틱 함수의 입력값은 학생이 시험에서 얻은 점수입니다. 만약 로지스틱 함수의 임계치를 60으로 설정한다면, 학생이 60점 이상이면 로지스틱 함수의 출력값이 1이 되어 합격으로 판단하고, 60점 미만이면 로지스틱 함수의 출력값이 0이 되어 불합격으로 판단할 수 있습니다.

이와 같이 로지스틱 함수는 이진 분류(binary classification) 문제에서 매우 유용하게 사용됩니다. 어떤 이메일이 스팸인지 아닌지를 판단하거나, 소비자가 제품을 구매할 것인지 아닌지를 예측하는 등의 문제에서 활용됩니다.



```python
def logistic(x):
    return 1 / (1 + np.exp(-x))

xx = np.linspace(-10, 10, 100)
plt.figure(figsize=(10,6))
plt.plot(xx, logistic(xx))
plt.title("로지스틱함수")
plt.xlabel("$x$")
plt.ylabel("$\sigma(x)$")
plt.show()
```

##로그함수

- e를 거듭제곱하여 특정한 값이 되는 수를 찾을 필요가 있는데 e를 몇 제곱해야 10이라는 값이 될까? e를 2.3025851 거듭제곱하면 10이 된다. e를 거듭제곱하여 특정한 수 a가 되도록 하는 수를 log a라 표기하고 로그라고 읽는다.

우리가 물건을 사거나 팔 때, 가격이 어떻게 변하는지 관찰하고 이를 예측하고자 할 때, 로그 함수를 사용할 수 있습니다. 로그 함수는 매우 큰 수나 작은 수를 다룰 때 유용한 함수입니다.

예를 들어, 1,000원짜리 과자를 100원씩 인상시키면, 10번의 인상 후 가격은 2,000원이 됩니다. 이때, 로그 함수를 적용하면 어떻게 될까요? 로그 함수의 밑을 2로 하여 적용한다면, 가격이 1,000원에서 2,000원으로 두 배가 되었으므로, 로그2(2) = 1입니다. 따라서 가격이 10번 인상될 때마다, 로그2(2)씩 증가한다는 것을 알 수 있습니다. 즉, 로그 함수는 수의 증가량이 일정할 때 이를 일정한 단위로 표현할 수 있도록 도와주는 함수라고 할 수 있습니다.

로그 함수는 데이터 분석에서 지수적으로 증가하는 양상을 보이는 데이터의 성장을 추적하는데 자주 사용됩니다. 예를 들어, 어떤 스타트업 회사의 매출액이 지수적으로 증가하고 있다면, 로그 함수를 사용하여 이 증가 양상을 더 잘 이해할 수 있습니다.

이와 같이 로그 함수는 매우 유용한 함수 중 하나입니다. 다양한 분야에서 활용되며, 데이터 분석에서도 자주 사용되는 함수입니다.

###로그함수의 성질1 : 로그함수는 곱하기를 더하기로 변환한다.
- log(x1,x2) = logx1 + logx2

당신은 집을 팔기 위해 집 가치를 계산하는 데 도와주기로 했어요. 하지만 당신이 계산해야 할 수치들이 너무 커서 계산하기가 어렵습니다. 그러던 중, 당신은 로그 함수를 이용하면 곱셈을 더하기로 바꿀 수 있다는 것을 알게 되었어요.

당신은 이제 집 가치를 계산할 때 로그 함수를 사용하여 곱셈 문제를 덧셈 문제로 바꿔서 계산할 수 있었어요. 예를 들어, 집 가치를 계산할 때 다음과 같은 수식을 사용할 수 있어요.

log(집 크기) + log(방 개수) + log(위치) = log(집 가치)


```python
import math

house_size = 100
num_rooms = 3
location = 2

# 로그 함수를 사용하여 곱셈 문제를 덧셈 문제로 변환
house_value = math.log(house_size) + math.log(num_rooms) + math.log(location)

# 다시 원래 값으로 변환
house_value = math.exp(house_value)

print("집 가치:", house_value)
```

###로그함수의 성질2 : 어떤 함수에 로그를 적용해도 함수의 최고점, 최저점의 위치는 변하지 않는다.

로그 함수를 적용하면 어떤 함수의 값을 바꿀 수 있어요. 하지만 그 함수의 최고점이나 최저점은 변하지 않아요. 최고점이란 함수 값이 가장 큰 지점을 의미하고, 최저점이란 함수 값이 가장 작은 지점을 의미해요.

예를 들어, 키가 150cm인 사람의 몸무게를 예측하는 함수를 생각해보자면, 이 함수의 최고점은 키와 몸무게의 관계에서 키가 커질수록 몸무게도 커지는 지점이 될 것입니다. 로그 함수를 적용해도 이 관계는 변하지 않아요. 로그 함수를 적용하기 전과 후에 최고점의 위치는 동일하게 유지되어요.

파이썬 코드 예시를 통해 이를 확인해보겠습니다.

```python
import math

# 키에 따른 몸무게 함수
def weight_func(height):
    return height ** 2 - 150 * height + 5000

# 로그 함수를 적용하기 전 최고점의 위치
print("로그 함수를 적용하기 전 최고점 위치:", -(-150) / (2 * 1))

# 로그 함수를 적용
def log_weight_func(height):
    return math.log(weight_func(height))

# 로그 함수를 적용한 후 최고점의 위치
print("로그 함수를 적용한 후 최고점 위치:", -(-150) / (2 * 1))
```

```python
def ff(x):
    return x**3 - 12*x + 20 * np.sin(x) + 7

xx = np.linspace(-4, 4, 300)
yy = ff(xx)
plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(xx, yy)
plt.axhline(1, c='r', ls="--")
plt.yticks([0, 1, 5, 10])
plt.ylim(-2, 15)
plt.title("$f(x)$")

plt.subplot(212)
plt.plot(xx, np.log(yy))
plt.axhline(0, c='r', ls="--")
plt.title("$log f(x)$")

plt.tight_layout()
plt.show()
```

###로그함수의 성질3 : 로그함수는 0부터 1사이의 작은 값을 확대시켜 보여준다.

당신은 미세먼지를 측정하는 일을 하게 되었어요. 미세먼지 수치가 매우 작아서 바로 이해하기 어려웠지만, 그래도 노력했어요. 하지만 더 이상 측정값이 작아지면 어떻게 해야할지 막막해졌어요.

그러던 중에 당신은 로그 함수를 이용하면 작은 값을 더 크게 보여줄 수 있다는 것을 알게 되었어요. 로그 함수는 0부터 1사이의 작은 값을 확대시켜 보여주기 때문에 매우 작은 측정값을 쉽게 다룰 수 있게 되었어요.

파이썬 코드로 예를 들어보면 다음과 같습니다.

```python
import math

# 미세먼지 수치 측정
dust = 0.0001

# 로그 함수를 사용하여 미세먼지 수치를 보여줌
dust_log = math.log(dust, 10)

print("미세먼지 수치:", dust)
print("로그 미세먼지 수치:", dust_log)
```
